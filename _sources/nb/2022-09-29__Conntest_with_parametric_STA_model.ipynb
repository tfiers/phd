## Params

```julia
p = get_params(
    duration = 10minutes,
    p_conn = 0.04,
    g_EE = 1,
    g_EI = 1,
    g_IE = 4,
    g_II = 4,
    ext_current = Normal(-0.5 * pA/√seconds, 5 * pA/√seconds),
    E_inh = -80 * mV,
    record_v = [1:40; 801:810],
);
```

## Load STA's

(They're precalculated).

```julia
@time begin
    q = @set (p.sim.general.duration = 1ms)
    s = VoltoMapSim.init_sim(q.sim)
    s = augment(s, q)
end;
```
