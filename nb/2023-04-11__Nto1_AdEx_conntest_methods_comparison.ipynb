## Why is AUC < 0.5

First, let's plot em

```julia
plot_dots_and_means(
    df, ycol;
    xcol = :N,
    xlabel = (xcol == :N ? "Number of inputs (N)" : string(xcol)),
    hylabel = string(ycol),
    color = C0,
    ax = nothing,
    kw...
) = begin
    clip_on = false
    isnothing(ax) && (fig, ax = plt.subplots())
    ax.set_xscale("log")
    x = df[:, xcol]
    y = df[:, ycol]
    plot(x, y, "."; clip_on, ms=8, alpha=0.6, color, ax)
    xu = unique(x)
    ax.set_xticks(xu, xu)
    ax.set_xlim(xu[1]/1.4, xu[end]*1.4)
    ax.minorticks_off()
    set(ax; xlabel, hylabel, kw...)
    dfm = combine(groupby(df, xcol), ycol => mean => ycol)
    ym = dfm[:, ycol]
    plot(xu, ym, "-"; clip_on, color, ax)
end

plot_AUC(df, method) = begin
    fig, ax = plt.subplots()
    ax.axhline(0.5, c="gray", lw=1, ls="--")
    plot_dots_and_means(
        df[df.method .== method, :],
        :AUC,
        title = method,
        ylim = [0, 1];
        ax
    )
end

methods = unique(df.method)
for m in methods
    plot_AUC(df, m)
end
```
