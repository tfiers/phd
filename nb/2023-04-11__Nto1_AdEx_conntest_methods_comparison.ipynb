```julia
using CSV
```

```julia
using DataFrames
```

```julia
df = CSV.read("../data/Nto1_AdEx.csv", DataFrame)
```

```julia
df[df.N .== 6500, :]
```

```julia
df2 = groupby(df, [:method, :N])
```

```julia
using Statistics
```

```julia
dfm = sort!(combine(df2, Between(:TPRₑ, :AUCᵢ) .=> mean), :N)
```

```julia
using WithFeedback
```

```julia
@withfb using PyPlot
```

```julia
function plotrates(m)

    df3 = df[df.method .== m, :]
    df3m = dfm[dfm.method .== m, :]

    # jitter(x) = x .* (0.95 .+ 0.1.*rand(length(x)))
    jitter(x) = x

    fig, ax = plt.subplots()
    plt.semilogx(jitter(df3.N), df3.TPR, ".", ms = 8)
    plt.semilogx(jitter(df3m.N), df3m.TPR_mean, "C0-")
    plt.semilogx(jitter(df3.N), df3.FPR, ".", ms = 3)

    Ns = unique(df.N)

    plt.xticks(Ns, Ns)
    plt.minorticks_off()

    plt.ylabel("Detection rate")
    plt.xlabel("Number of inputs (N)")
    plt.title(m)
end

plotrates("fit_upstroke")
plotrates("STA_height")
plotrates("STA_corr_2pass")
;
```
