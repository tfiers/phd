Getting intuition for partial correlation i.e. inverting correlation matrix

```python
from tfiers.nb import *
```

```python
S = [
  [1.0, 0.0, 0.0, 0.0],
  [-0.8, 1.0, 0.0, 0.0],
  [0.6, 0.8, 1.0, 0.0],
  [0.4, 0.6, 0.8, 1.0],
]

# make symmetric
S = np.array(S)
S += np.tril(S, k=-1).T
    # Select lower triangular from one below diagonal, set rest to 0, transpose, add

set_float_print_format('+.1f')

conn = -np.linalg.inv(S)
```

```python
def plot(matrix, **kwargs):
    plt.imshow(matrix, **kwargs)
    plt.xticks([])
    plt.yticks([])
```

```python
plot(S, vmin=-1, vmax=1, cmap='Blues');
```

```python
plot(conn, cmap='Blues');
```
